### Scipy.signal.find_peaks Test

https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.find_peaks.html

```python
import matplotlib.pyplot as plt
from scipy.misc import electrocardiogram
from scipy.signal import find_peaks
x = electrocardiogram()[2000:4000]
height = -0.5

peaks, properties = find_peaks(x, height=height)
plt.plot(x)
plt.plot(peaks, x[peaks], "x")
plt.plot(np.zeros_like(x)+height, "--", color="gray")
plt.show()
```

```python
plt.plot(properties['peak_heights'])
```

```python
border = np.sin(np.linspace(0, 3 * np.pi, x.size))
peaks, properties = find_peaks(x, height=(-border, border))
plt.plot(x)
plt.plot(-border, "--", color="gray")
plt.plot(border, ":", color="gray")
plt.plot(peaks, x[peaks], "x")
plt.show()
```

```python
plt.plot(properties['peak_heights'])
```

```python
peaks, properties = find_peaks(x, distance=100)
np.diff(peaks)
plt.plot(x)
plt.plot(peaks, x[peaks], "x")
plt.show()
```
